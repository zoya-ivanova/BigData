Подсчитайте среднее значение и дисперсию по признаку ”price”.



```python
import pandas as pd
import numpy as np
```

```python
#Подгружаем данные
data = pd.read_csv('/AB_NYC_2019.csv')

# Смотрим датафрейм
data.head(10)
```

```python
# Проверяем тип данных
data.info()
```

```python
# размеры таблицы по двум осям
data.shape
```

```python
# проверяем пустые значения
data.isna().sum()
```

```python
# Проверяем наличие дубликатов
data.duplicated().sum()
```

```python
# % пропущенных значений в каждом столбце
for col in data.columns:
  missing_values = np.mean(data[col].isnull())
  print('{} - {}%'.format(col, round(missing_values * 100)))
```

```python
# заполняем значения колонок, где 10% и более пропущенных значений,
# предварительно заменив тип данных,
# дату средним нельзя заполнить, заполняем - NaN
data['last_review'] = data['last_review'].astype('str')
data['last_review'] = data['last_review'].fillna('NaN')
```

```python
# заполняем значения средним (медианой), т.к оно числовое
mediana = data['calculated_host_listings_count'].median()
data['calculated_host_listings_count'] = data['calculated_host_listings_count'].fillna(mediana)
```

```python
# видим в колонке 'price' - нулевые значения
data.sort_values (by='price')
```

```python
# заполняем значения средним (медианой)
mediana = data['price'].median()
data['price'] = data['price'].fillna(mediana)
```

```python
# Подсчет среднего значения по признаку 'price'
mean_price = data['price'].mean()
print(f"Среднее значение цены по признаку 'price': {mean_price:.2f}")

# Подсчет дисперсии по признаку "price"
variance_price = data['price'].var()
print("Дисперсия по признаку 'price':", variance_price)
```
